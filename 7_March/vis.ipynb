```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,3,4,7],"go")
plt.ylabel("ABC")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"rs")  # x values followed by y values  Format string, first for color, 'o' means dot, '-' means continuous line, '--' means dotted line, 's' for square box, '^' for triangular box
plt.axis((0, 6, 0, 20))      #[xmin, xmax, ymin, ymax]  # maximum values
plt.xlabel('some numbers')
plt.ylabel('square some numbers')
plt.show()
```

```python
t=np.arange(0.,5.,0.2)
print(t)
plt.plot(t,t,"ro",t,t**2,"gs",t,t**3,"b^")
plt.show()
```
