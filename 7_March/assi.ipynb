```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming you've already loaded the dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("diabetes.csv", header=None, names=col_names)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot for pregnant vs label
pregnant_label_counts = pima.groupby(['pregnant', 'label']).size().unstack()
pregnant_label_counts.plot(kind='bar', stacked=True, ax=axs[0, 0])
axs[0, 0].set_title('Pregnant vs Label')
axs[0, 0].set_xlabel('Pregnant')
axs[0, 0].set_ylabel('Count')
axs[0, 0].legend(title='Label', loc='upper right')

# Plot for glucose vs label
glucose_label_counts = pima.groupby(['glucose', 'label']).size().unstack()
glucose_label_counts.plot(kind='bar', stacked=True, ax=axs[0, 1])
axs[0, 1].set_title('Glucose vs Label')
axs[0, 1].set_xlabel('Glucose')
axs[0, 1].set_ylabel('Count')
axs[0, 1].legend(title='Label', loc='upper right')

# Plot for bmi vs label
bmi_label_counts = pima.groupby(['bmi', 'label']).size().unstack()
bmi_label_counts.plot(kind='bar', stacked=True, ax=axs[1, 0])
axs[1, 0].set_title('BMI vs Label')
axs[1, 0].set_xlabel('BMI')
axs[1, 0].set_ylabel('Count')
axs[1, 0].legend(title='Label', loc='upper right')

# Plot for bp vs label
bp_label_counts = pima.groupby(['bp', 'label']).size().unstack()
bp_label_counts.plot(kind='bar', stacked=True, ax=axs[1, 1])
axs[1, 1].set_title('Blood Pressure (BP) vs Label')
axs[1, 1].set_xlabel('BP')
axs[1, 1].set_ylabel('Count')
axs[1, 1].legend(title='Label', loc='upper right')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you've already loaded the dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("diabetes.csv", header=None, names=col_names)

# Filter dataset for instances where the label is 1 (indicating diabetes)
diabetes_age = pima[pima['label'] == 1]['age']

# Plot histogram
plt.hist(diabetes_age, bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Individuals with Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you've already loaded the dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("diabetes.csv", header=None, names=col_names)

# Plot boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='label', y='glucose', data=pima)
plt.title('Glucose Level of Patients with and without Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Glucose Level')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()
```
