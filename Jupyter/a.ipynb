```python
#6
def y(a):
    if (a % 4 == 0 and a % 100 != 0) or (a % 400 == 0):
        return True
    else:
        return False

def check():
    try:
        a = int(input("Enter a a: "))
        if y(a):
            print(f"{a} is a leap.")
        else:
            print(f"{a} is not a leap.")
    
    except ValueError:
        print("Invalid input. Please enter a valid a.")

if __name__ == "__main__":
    check()
```

```python
#7
def grade(m):
    if 90 <= m <= 100:
        return 'O'
    elif 80 <= m < 90:
        return 'E'
    elif 70 <= m < 80:
        return 'A'
    elif 60 <= m < 70:
        return 'B'
    elif 50 <= m < 60:
        return 'C'
    elif 40 <= m < 50:
        return 'D'
    elif 0 <= m < 40:
        return 'F'
    else:
        return 'Invalid Input'

def prin():
    try:
        m = float(input("Enter total marks secured: "))
        if 0 <= m <= 100:
            g = grade(m)
            print(f"The grade for {m} marks is: {g}")
        else:
            print("Invalid input. Total marks should be between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")

if __name__ == "__main__":
    prin()
```

```python
#9
def sum():
    b=int(input("Enter a no. :- "))
    try:
        n=0
        a=b
        while(a):
            n = n+ (a % 10)
            a=a // 10
        print(f"The sum of digits of {b} are {n}")
            
    except ValueError:
        print("Invalid input")
        
if __name__ == "__main__":
    sum()
            
            
```

```python
#10
def prime(b):
    if b < 2:
        return False
    for i in range(2, int(b**0.5) + 1):
        if b % i == 0:
            return False
    return True

def check_prime():
    try:
        b = int(input("Enter a number: "))
        if prime(b):
            print(f"{b} is a prime number.")
        else:
            print(f"{b} is not a prime number.")

    except ValueError:
        print("Invalid input")

if __name__ == "__main__":
    check_prime()
```

```python
#11
def perf():
    a=int(input("Enter a no. :-")) 
    x=0
    for i in range(1,a):
        if (a % i) == 0:
            x = x + i
                      
    if x==a:
        print(f"{a} is a perfect.")
    else:
        print(f"{a} is not perfect.")
        
if __name__ == "__main__":
    perf()
```

```python
#12
def fib(n):
    seq=[0,1]
    
    for i in range(2,n):
        x=seq[i-1]+seq[i-2]
        seq.append(x)
        
    return seq[:n]

def find():
    n=int(input("Enter a no. :-"))
    
    r=fib(n)
    print(f"The seq{r}")
    
if __name__ == "__main__":
    find()
```

```python
#13
def arm(b):
    
    x = str(b)
    num_digits = len(x)
    
    armstrong_sum = sum(int(digit) ** num_digits for digit in x if digit.isdigit())
    return armstrong_sum == b

if __name__ == "__main__":
    try:
        
        b = int(input("Enter a no to check if it's an Armstrong :- "))

       
        if arm(b):
            print(f"{b} is an Armstrong.")
        else:
            print(f"{b} is not an Armstrong.")

    except ValueError:
        print("Invalid input. Please enter a valid a.")
```

```python
#14
import random

def game():
    print("Welcome to Game")
    print("A no. is generate from 1 to 100(GUESS)")
    
    a=random.randint(1,100)
    
    while True:
        try:
            b=int(input("Enter Your Guess :"))
            if b==a:
                 print("Congratulations! You guessed the correct number.")
                 break
            
            elif b < a:
                print("Too low! Try again.")
            else:
                 print("Too high! Try again.")
                 
        
        except ValueError:
            print("Invalid input")
            

if __name__ == "__main__":
    game()    
```
